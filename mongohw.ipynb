```python
#Write a MongoDB query to display average number of awards won by a movie (use aggregate function with $avg operator).


df = pd.DataFrame(list(
    db.movies.aggregate([
        {"$group": {"_id": None, "averageAwards": {"$avg": "$awards.wins"}}},
        {"$project": {"_id": 0, "averageAwards": 1}}
    ])
))

df
```

```python
#Write a MongoDB query to display most awards won by a movie (use aggregate function with $max operator). Comments Collection General Queries
df = pd.DataFrame(list(
    db.movies.aggregate([
        {"$group": {"_id": None, "mostAwardsWon": {"$max": "$awards.wins"}}},
        {"$project": {"_id": 0, "mostAwardsWon": 1}}
    ])
))

df
```

```python
#Write a MongoDB query to display the total number of documents in the collection movies.
df = pd.DataFrame(list(
    db.movies.aggregate([
        {"$group": {"_id": None, "totalDocuments": {"$sum": 1}}},
        {"$project": {"_id": 0, "totalDocuments": 1}}
    ])
))

df
```

```python
#Write a MongoDB query to display the total number of users by name (use db.collection.distinct(field, query, options) and length) in the collection movies.
df = pd.DataFrame(list(
    db.movies.aggregate([
        {"$group": {"_id": None, "totalDocuments": {"$sum": 1}}},
        {"$project": {"_id": 0, "totalDocuments": 1}}
    ])
))

df
```
